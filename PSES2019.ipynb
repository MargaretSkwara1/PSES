```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
pd.set_option('display.max_colwidth', -1)
```

```python
subset1_url = 'https://www.canada.ca/content/dam/tbs-sct/documents/datasets/2019_PSES_SAFF_%20subset-1_Sous-ensemble-1.csv'
df = pd.read_csv(subset1_url, encoding = 'latin1')
df.columns
df.shape
df.head(3)
```

```python
df['DESCRIP_E'].nunique()
print(df['DESCRIP_E'].unique().tolist())
```

## Pick list of Agencies to compare rank

```python
filters = ['Public Service', 'National Energy Board', 'Environment and Climate Change Canada', 'Statistics Canada', \
           'Natural Resources Canada', 'Western Economic Diversification Canada']

int_columns = ['SURVEYR', 'DESCRIP_E', 'TITLE_E', 'SCORE100','INDICATORENG', 'DEPT_E', 'ANSCOUNT']
df_filtered = df[df['DESCRIP_E'].isin(filters)].copy()
df_sorted = df_filtered[int_columns]
df_sorted.loc[:, 'rank'] = df_sorted.groupby(['SURVEYR', 'TITLE_E'])['SCORE100'].rank(ascending=False)
df_sorted.shape
df_sorted.head(12)
df_sorted.to_csv('test.csv', index = False)
```

## All Federal Agencies ranked

```python
df_filtered_all = df
df_sorted_all = df_filtered_all[int_columns]
df_sorted_all.loc[:, 'rank'] = df_sorted_all.groupby(['SURVEYR', 'TITLE_E'])['SCORE100'].rank(ascending=False).copy()
df_sorted_all = df_sorted_all.sort_values(by = ['rank'])
df_sorted_all.head()
df_sorted_all.to_csv('all_agencies_ranked.csv', index = False)
```

## CER only

```python
cer_only = df[df['DESCRIP_E'].str.contains('National Energy Board')]
cer_only = cer_only[int_columns]
cer_only['SURVEYR'] = pd.to_datetime(cer_only['SURVEYR'], format = '%Y')
# cer_only = cer_only.sort_values(by = ['TITLE_E', 'SURVEYR'])
cer_only['change'] = cer_only.sort_values(by = ['TITLE_E', 'SURVEYR']).groupby(['TITLE_E'])['SCORE100'].diff()
```
